```python
#----- ì‹¤ìŠµ1
import csv
f = open('C:/Users/mink9/OneDrive/ë°”íƒ• í™”ë©´/ì˜¤í”ˆì†ŒìŠ¤SWìž…ë¬¸/temp2011-2020.csv','r',encoding = 'cp949')
data = csv.reader(f)
next(data)
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
high=[]
low=[]
for row in data:
    if row[0].split('-')[1]=='02' and row[0].split('-')[2]=='05':
        high.append(float(row[-1]))
        low.append(float(row[-2]))
f.close()

plt.title("2011~2020 Highest, Lowest temperature Febrary 5")
plt.plot(year,high,label='Highest Temperature')
plt.plot(year,low,label='Lowest Temperature')
plt.legend()
plt.show()
```

```python
#----- ì‹¤ìŠµ2
import csv
f = open('C:/Users/mink9/OneDrive/ë°”íƒ• í™”ë©´/ì˜¤í”ˆì†ŒìŠ¤SWìž…ë¬¸/temp2011-2020.csv','r',encoding = 'cp949')
data = csv.reader(f)
next(data)
jan=[]
aug=[]
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        elif month == '08':
            aug.append(float(row[-1]))
f.close()

plt.title("2011~2020 Highest temperature in January, August")
plt.hist(jan,bins=100,label='Jan')
plt.hist(aug,bins=100,label='Aug',color='r')
plt.legend()
plt.show()
```
